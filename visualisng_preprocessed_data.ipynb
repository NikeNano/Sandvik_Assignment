## Label
Convert the string label to a integer. 0 represent female and 1 represent male. Some of the labels are in german and spanish. Only select the observations with a label representing Female or Male.

```python
df["label"].unique()
```

```python
# Masculino==male, portugise
# Weiblich==female german
label_mapping={"Male":1,"male":1,"Masculino":1,"Female":0,"female":0,"Weiblich":0}
df_selected=df.replace({'label':label_mapping})
df_selected=df_selected[(df_selected['label']==0)|(df_selected['label']==1)]
df_selected.label.unique()
```

## Dialect

Adjust the dialect mapping. 

```python
df_selected[['dialect']]=df_selected[['dialect']].fillna("Other")
df_selected['dialect'].unique()
```

# PCA on the data

Often hard to see natural seperation in the data over multiple variables. Usina a PCA and selecting the first two principal components offer a possible way to see if there is any clear seperation in the data. 

First step is to normalise the data 

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(df_selected[continues_columns])
df_selected_scaled = pd.DataFrame(x_scaled)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca_data = pca.fit_transform(df_selected_scaled)
```

```python
explained_ratio=pca.explained_variance_ratio_
print("The first component {}% of the variance".format(explained_ratio[0]))
print("The second component {}% of the variance".format(explained_ratio[1]))
```

```python
df_pca = pd.DataFrame(data = pca_data,columns=['pca_comp1','pca_comp2','pca_comp3','pca_comp4','pca_comp5','pca_comp6','pca_comp7','pca_comp8','pca_comp9','pca_comp10'])
df_pca['label']=df_selected['label'].values
```

Very hard to see and seperation of the data due to that is overlayin each other. One way to try to takle this to sample the data and then look if it is overlayin in the whole data. 

```python
sns.lmplot( x="pca_comp1", y="pca_comp2",
  data=df_pca, 
  fit_reg=False, 
  hue='label', # color by cluster
  legend=True,
  scatter_kws={"s": 80,'alpha': 0.5}) # specify the point size
```

Looking at the sampled plot below where 1% of the data is sampled it is hard to see any seperation of the data. 

```python
df_pca_sample=df_pca.sample(frac=0.01)
sns.lmplot( x="pca_comp1", y="pca_comp2",
  data=df_pca_sample, 
  fit_reg=False, 
  hue='label', # color by cluster
  legend=True,
  scatter_kws={"s": 80,'alpha': 0.5}) # specify the point size
```
